```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "nitrosocke/Ghibli-Diffusion"  # Ghibli-style Stable Diffusion model

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")  # Move model to GPU
```

```python
prompt = "a peaceful Ghibli-style village with lush greenery and warm sunset lighting"
image = pipe(prompt).images[0]
image.show()
```

```python
from diffusers import StableDiffusionImg2ImgPipeline
import torch
from PIL import Image

# Load the correct pipeline
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "nitrosocke/Ghibli-Diffusion", torch_dtype=torch.float16
).to("cuda")

# Define the prompt
prompt = "A Studio Ghibli-style illustration, anime aesthetic, soft colors, detailed fantasy setting"

# Resize image to 512x512 for best results
image = image.resize((512, 512))

# Apply the Ghibli style
output = pipe(prompt=prompt, image=image, strength=0.6, guidance_scale=7.5)
styled_image = output.images[0]

# Save and display the output image
styled_image.save("ghibli_styled.png")
display(styled_image)
```

```python
from diffusers import StableDiffusionImg2ImgPipeline
import torch

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "nitrosocke/Ghibli-Diffusion", torch_dtype=torch.float16
).to("cuda")
```

```python
# Define a detailed Ghibli-style prompt
prompt = """Ghibli style image of given image. change background to stunning water. and make tree branches green.
"""
output = pipe(
    prompt=prompt,
    image=image,
    strength=0.2,  #increase for better stylization
    num_inference_steps=200,  #for better details
    guidance_scale=9,  # Higher prompt adherence
    eta=0.5,  #smoother results
    negative_prompt="blurry, distorted face, low quality, oversaturated, extra limbs, missing details, unnatural colors"
)

styled_image = output.images[0]

styled_image.save("ghibli_styled1.png")
display(styled_image)
```
